```python
import random
products = ['chair','table','laptop','mobile phone','charger']
region = ['north','south','east','west']
date = pd.date_range('01-01-2025','23-07-2025')

data = {'Order ID':range(1,1001),
        'Date':random.choices(date,k=1000),
    'Region Name':random.choices(region,k=1000),  #k=1000; no of times it has to be dispalyed
       'Product Name':random.choices(products,k=1000),
       'Sales':np.random.randint(100,10000,1000),
       'Quantity':np.random.randint(1,10,1000),
       'Profit':np.random.randint(-1000,10000,1000)}

table = pd.DataFrame(data)
table
```
