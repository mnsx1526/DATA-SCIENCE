```python
for i in cat_col:
    print(f'\n\nTOP 10 ANALYSIS BY {i}\n\n')
    display(df[i].value_counts().head())
```

```python
# Top 10 Analysis For Categorical Columns in %

for i in cat_col:
  print(f'\n\nTop 10 Analysis by {i}\n')
  top_10 = df[i].value_counts().head(10)

  ans = (top_10/top_10.sum())*100

  display(ans.round(2))
```

```python
#Bottom 10 Analysis For Categorical Columns

for i in cat_col:
  print(f'\n\nBottom 10 Analysis by {i}\n')
  bottom_10 = df[i].value_counts().tail(10)
  display(bottom_10)
```

```python
df['car_name'].value_counts().head(10).plot(kind = 'pie',autopct = '%.2f%%')
```

```python
import matplotlib.pyplot as plt
def graphplot(col_name):
    x = df['car_name'].value_counts().head(10).index
    y = df['car_name'].value_counts().head(10).values

    plt.figure(figsize = (8,5))
    plt.title(f'Top 10 {col_name} By Inventory',fontsize = 20,color = 'Brown')
  # ax = plt.bar(x,y, color = 'red')

    ax = plt.bar(x,y, color = sns.color_palette('pink',10))
    plt.bar_label(ax)
    plt.xticks(rotation = 45)
    plt.show()
```

```python
for i in cat_col:
    graphplot(i)
```

```python
import numpy as np

def plot_hist(col_name,bin_size = 100):
  plt.title(f'{col_name} Distribution Analysis',fontsize = 20,color = 'pink' )
  plt.hist(df[col_name],bins = np.linspace(df[col_name].min(),df[col_name].max(),bin_size))

  plt.show()
```

```python
plot_hist('engine')
```

```python
for i in num_col:
    plot_hist(i)
```
