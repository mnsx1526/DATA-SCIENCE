```python
# Internet Required
df_titanic = sns.load_dataset('titanic')
df_titanic.sample()
```

```python
import matplotlib.pyplot as plt
# to use seaborn we have to call matplotlib first
plt.title('TITANIC SURVIVAL ANALYSIS BASED ON GENDER')
sns.countplot(data = df_titanic, x = 'sex', hue = 'alive', palette = sns.color_palette('magma',2))  # hue: further divide: based on categorical(object)
plt.show()
```

```python
sns.barplot(data = df_titanic, x='class', y='fare',hue='alive',errorbar=('ci',0))
plt.grid(color='r',alpha = 0.1)
# the line in b/w the bars is known as CI  : confidence interval'
# estimator is the min, max , mean by default it is mean
plt.show()
# Confidence Interval(CI)/error bar: 95%
```

```python
# ************* V IMP ************************
# sns.boxplot: To detect Outliers and to check data distribution
# outliers - jhoote data points
# Range: Min to max(19 - 22): 35, 28, 45(Outliers: fliers)
# Q2 : quantile : 50%
# Q3 : 75%
# Q1 : 25%
# UW = Q3 +1.5*IQR
# LW = Q1 +1.5*IQR
```

```python
sns.boxplot(data = df_titanic, x = 'fare', y='class', hue='sex',orient = 'h')
# the line in the middle of the box indicates the average
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data = df_titanic, x = 'fare', y='class', hue='sex',orient = 'h')
# the line in the middle of the box indicates the average
plt.show()
```
